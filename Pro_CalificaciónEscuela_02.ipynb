```python
CalificacionEvaluacion = int(input("Escribe la calificación de la evaluación"))
print(CalificacionEvaluacion)

PorcentajeEvaluacion = 50 * CalificacionEvaluacion /100

print (f"Tu calificación de la evaluación es: " + str(PorcentajeEvaluacion))
```

```python
CalificacionTrabajoFinal = int(input("Escribe la calificación del Trabajo final"))
print(CalificacionTrabajoFinal)

PorcentajeTrabajoFinal = 30 * CalificacionTrabajoFinal /100

print (f"Tu calificación del Trabajo Final es: " + str(PorcentajeTrabajoFinal))
```

```python
CalificacionTareas = int(input("Escribe la calificación de tus Tareas"))
print(CalificacionTareas)

PorcentajeTareas = 20 * CalificacionTareas / 100

print (f"Tu calificación de Tareas es: " + str(PorcentajeTareas))
```

```python
CalificacionFinal = PorcentajeEvaluacion + PorcentajeTrabajoFinal + PorcentajeTareas
print(f"Tu calificación del semestre es: " + str(CalificacionFinal))
```
